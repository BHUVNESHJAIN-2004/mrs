```python
# convert string list into integer list
# literal_eval do that for us 

import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
import ast


def convert2(obj):

    L = []
    count = 0
    for i in ast.literal_eval(obj):

        if count<3:
            L.append(i["name"])
            count+=1
        else:
            break


    return L
        
    
```

### Fetch director in crew

```python
def fetch_director(obj):

    L = []

    for i in ast.literal_eval(obj):

        if i['job']=='Director':
            L.append(i['name'])

    return L
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

### Now in columns genres,keywords , cast and crew we remove space between them 
so we diffrentiate between two names 
(Sam Worthington convert into SamWorthington)

```python
df['genres'] = df['genres'].apply(lambda x : [i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x : [i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

## Now create a new column tags which contains all column values (genres	keywords	overview	cast	crew)

```python
df['tags'] = df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
# drop all columns 

df.drop(columns=['overview','cast','crew','genres','keywords'],inplace=True)
```

# Now convert list into Strings

```python
df['tags'] = df['tags'].apply(lambda x :" ".join(x))
```

```python
df.head()
```

## Convert tage into lower case

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head(1)
```

# now we perform stemming so our words love loving lover treated as one word

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

### helper function for converting text

```python
def stem_text(text):

    L = []

    for i in text.split():
        L.append(ps.stem(i))

    return " ".join(L)
```

```python
df['tags'] = df['tags'].apply(stem_text)
```

# Perform vectorization on tags so we find similarity between movies

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()[0:500]
```

## Now find distance of every vector from all other vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

# Now create a function recommanded in which if user provide a movie name then we return it 5 similar movies

```python
def recommend(movie):

    # find index of movie
    index = df[df['title']==movie].index[0]

    # sort all distances for this movie index
    # here first we combine all movies with their indexes and then we combine all of them in a list after we sort
    # here key is fucntion which is used to sort basis on distance not index
    distances = similarity[index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]

    # now fetch top 5 movies

    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
